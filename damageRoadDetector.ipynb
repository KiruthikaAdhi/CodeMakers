```python
from PIL import Image
```

```python
preDisasterRoad = "classify_output/channelviewPreRoad1.jpeg"
postDisasterRoad = "classify_output/channelviewPostRoad1.jpeg"

preDsrRoadImg = Image.open(preDisasterRoad)
postDsrRoadImg = Image.open(postDisasterRoad)
 
preX, preY = preDsrRoadImg.size
postX, postY = postDsrRoadImg.size

print(preX, preY)
print(postX, postY)

#print(preDsrRoadImg[x,y])
```

```python
damageRoad = "damageRoad.jpeg"
damageRoadImg = Image.new('RGB', (preX, preY), (0, 0, 0))
damageRoadImg.save(damageRoad)
```

```python
damageRoadPixels = damageRoadImg.load()        
for x in range(preX):
    for y in range(preY):
        prePixel = preDsrRoadImg.getpixel((x, y))
        postPixel = postDsrRoadImg.getpixel((x, y))
        if(prePixel == (255,255,255)):
            print(prePixel, postPixel)
            if(postPixel == (0, 0, 0)):
                #print("Damage")
                damageRoadPixels[x, y] = (255, 0, 0)
            else:
                damageRoadPixels[x, y] = (0, 0, 0)
                
        else:
            damageRoadPixels[x,y] = (0, 0, 0)
damageRoadImg.save(damageRoad)

                
```

```python
def overlay(postDisasterImg, damagedRoadImg):
    overlay = "overlayDamagedRoad.jpeg"
    overlayedRoadImg = Image.open( "post_disaster/channelview.tiff")
    overlayedRoadImg = overlayedRoadImg.resize((512,512),Image.ANTIALIAS)
    px, py = damagedRoadImg.size
    overlayedRoadPixels = overlayedRoadImg.load()
    for x in range(0, px) :
        for y in range(0, py):
            d = damageRoadImg.getpixel((x, y))
            if( d == (255, 0, 0) ):
                #print("damage")
                overlayedRoadPixels[x+4 , y+4] = (255, 0, 0)    
    overlayedRoadImg.save(overlay)
                    
postDisaster = "post_disaster/channelview.tiff"
postDisasterImg = Image.open(postDisaster)
postDisasterImg = postDisasterImg.resize((512,512),Image.ANTIALIAS)
damageRoad = "damageRoad.jpeg"
damagedRoadImg = Image.open(damageRoad)
overlay(postDisasterImg, damagedRoadImg)
```
