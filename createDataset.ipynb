```python
from PIL import Image
from os import listdir
import random
l = listdir("data/train_input")
print(len(l))

fileName = "trainData.csv"
dataFile = open(fileName, 'w')

inputImage = Image.open("data/train_input/" + l[0])
inputImageXSize, inputImageYSize = inputImage.size
print(inputImageXSize)
print(inputImageYSize)
inputImagePixels = inputImage.load()
print(inputImagePixels)

# for i in range(inputImageXSize) : 
#     #for j in range(inputImageYSize) : 
#     print(inputImagePixels[i,j])
```

```python
def writeDataFile(inputImgPath, outputImgPath, inputImgFiles, outputImgFiles, fileName):
    dataFile = open(fileName, 'w')
    rectSize = 5
    linesLimit = 90000
    linesCount = 0
    linesCountPerImg = 0
    linesLimitPerImg = (linesLimit // len(inputImgFiles)) + 1
    print(linesLimitPerImg )
    for i in range(len(inputImgFiles)):
        linesCountPerImg = 0
        print(inputImgFiles[i])
        ip = Image.open(inputImgPath + '/' + inputImgFiles[i])
        ipX, ipY = ip.size
        
        op = Image.open(outputImgPath + '/' + outputImgFiles[i])
        opX, opY = op.size
        opPixels = op.load()
#         for i in range(opX):
#             for j in range(opY):
#                 print(opPixels[i,j])
        road = []
        nonRoad = []
        
        for x in range(rectSize//2, ipX - (rectSize//2)):
            for y in range(rectSize//2, ipY - (rectSize//2)):
                isRoad = opPixels[x , y]
                #print("isRoad : %d", isRoad , " opPixels ", opPixels[x,y] )
                if(isRoad):
                    road.append((x,y))
                else:
                    nonRoad.append((x,y))
        #print(road)
        #print(nonRoad)
        random.shuffle(road)
        random.shuffle(nonRoad)
        
        for m in range(len(road)):
            print("Road : ",i)
            if(linesCountPerImg >= linesLimitPerImg):
                break
            if((m*2) + 1 >= len(nonRoad)) :
                break
                
            x = road[m][0]
            y = road[m][1]
            
            subRect = ( x - (rectSize//2), y - (rectSize//2), x + (rectSize//2) + 1, y + (rectSize//2) + 1)
            subImg = ip.crop(subRect).load()
            line = ''
            for k in range(rectSize):
                for j in range(rectSize):
                    line += str(subImg[k,j][0]) + ','
                    line += str(subImg[k,j][1]) + ','
                    line += str(subImg[k,j][2]) + ','
            line += str(1) + '\n'
            linesCount += 1
            linesCountPerImg += 1
            dataFile.write(line)
            
            for n in range(2):
                print("Non Road : ",i)
                x = nonRoad[(m*2) + n][0]
                y = nonRoad[(m*2) + n][1]
                subRect = ( x - (rectSize//2), y - (rectSize//2), x + (rectSize//2) + 1, y + (rectSize//2) + 1)
                subImg = ip.crop(subRect).load()
                line = ''
                for k in range(rectSize):
                    for j in range(rectSize):
                        line += str(subImg[k,j][0]) + ','
                        line += str(subImg[k,j][1]) + ','
                        line += str(subImg[k,j][2]) + ','
                line += str(0) + '\n'
                linesCount += 1
                linesCountPerImg += 1
                dataFile.write(line)
        print(linesCount)
        
inputImgPath = "data/train_input"
inputImgFiles = listdir(inputImgPath)
inputImgFiles = inputImgFiles
print(inputImgFiles)
outputImgPath = "data/train_output"
outputImgFiles = listdir(outputImgPath)
outputImgFiles = outputImgFiles
inputImgFiles.sort()
outputImgFiles.sort()
print(inputImgFiles)
print(outputImgFiles)

if(len(inputImgFiles) != len(outputImgFiles)):
    print("Dataset doesnot match")
flag = 1    
for i in range(len(inputImgFiles)):
    print(inputImgFiles[i][:-1] , "," , outputImgFiles[i] )
    if(inputImgFiles[i][:-1] != outputImgFiles[i]):    
        print("Name does not match")
        flag = 0

if(flag):        
    writeDataFile(inputImgPath, outputImgPath, inputImgFiles, outputImgFiles, fileName)

        
    
```
